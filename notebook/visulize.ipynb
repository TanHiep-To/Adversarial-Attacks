```python
import sys
sys.path.append('../')
import numpy as np
from datetime import datetime 
%matplotlib inline
import matplotlib.pyplot as plt
from utils.dataset import Dataset

config = {
    'data_dir': '../data/images/',
}
```

#### Load dataset

```python
dataset = Dataset(config['data_dir'])
dataset.load_data()
dataset.load_labels()
```

### Visualize the data

```python
plt.figure(figsize=(10, 10))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    image = dataset[i]
    plt.imshow(image)
    label = dataset.labels[i]
    plt.title(f'{label}')
    plt.axis('off')
```
